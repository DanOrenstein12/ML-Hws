```python
import numpy as np
import csv
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import f1_score
from sklearn.model_selection import StratifiedKFold

def read_data():
    data = []
    with open('transfusion.data') as csvfile:
        reader = csv.reader(csvfile)
        headers = next(reader) # take the header out
        for row in reader: # each row is a list
            data.append(row)
    data  = np.array(data, dtype = np.int32)
    X = data[:,:-1]
    y = data[:,-1]

    return X, y

X, y = read_data()

def nested_cross_validation(X, y, folds, parameters):
    skf = StratifiedKFold(n_splits=folds, shuffle=True)
    results = []
    for train_index, test_index in skf.split(X, y):
        X_train, X_test = X[train_index],  X[test_index]
        y_train, y_test = y[train_index],  y[test_index]
        best_param = 
        cross_validate(X_train, y_train, parameters, folds-1)
        model = LogisticRegression(C=best_param)
        model.fit(X_train, y_train) # training
        y_pred = model.predict(X_test) # predicting
        current_f1_score = f1_score(y_test, y_pred)
        results.append(current_f1_score)
    mean = np.mean(results)
    st_dev = np.std(results)
    print(mean, st_dev)










def cross_validate(X_test, y_test, parameters, folds):
    eval = [0,0,0,0]
    skf_2 = StratifiedKFold(n_splits=folds)

    for train_index, test_index in skf_2.split(X_test, y_test):
        X_train, X_validate = X_test[train_index],  X_test[test_index]
        y_train, y_validate = y_test[train_index],  y_test[test_index]
        for i in range(len(parameters)):
            model = LogisticRegression(C=parameters[i])
            model.fit(X_train, y_train)
            y_pred = model.predict(X_validate)
            current_f1_score = f1_score(y_validate, y_pred)
            eval[i] = eval[i] + current_f1_score

    print(eval)
    print(parameters[eval.index(max(eval))])
    return parameters[eval.index(max(eval))]


nested_cross_validation(X, y, 5, [.1, 1, 10, 100])
```
